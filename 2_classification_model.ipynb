```python
# Model Training
import joblib
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# --- 1. Load train/test data ---
X_train = joblib.load('../data/X_train_sparse.pkl')
y_train = joblib.load('../data/y_train_labels.pkl')

# --- 2. Initialize models ---
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'XGBoost': XGBClassifier(eval_metric='mlogloss', random_state=42, use_label_encoder=False)
}

# --- 3. Train each model and save ---
for name, model in models.items():
    print(f"\n>>> Train of the model : {name}")
    model.fit(X_train, y_train)
    joblib.dump(model, f'../models/{name.lower().replace(" ", "_")}.pkl')
    print(f"Model {name} saved.")

# Model Evaluation
import joblib
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Load test data and label encoder ---
X_test = joblib.load('../data/X_test_sparse.pkl')
y_test = joblib.load('../data/y_test_labels.pkl')
label_encoder = joblib.load('../models/label_encoder.pkl')

# --- 2. Load saved models ---
model_names = ['logistic_regression', 'random_forest', 'xgboost']
models = {name: joblib.load(f'../models/{name}.pkl') for name in model_names}

# --- 3. Function to display confusion matrix ---
def plot_confusion_matrix(y_true, y_pred, classes, title='Matrice de confusion'):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Prédit')
    plt.ylabel('Réel')
    plt.title(title)
    plt.xticks(rotation=90)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# --- 4. Evaluate each model ---
for name, model in models.items():
    print(f"\n>>> Model evaluation : {name.replace('_', ' ').title()}")
    y_pred = model.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)
    f1_macro = f1_score(y_test, y_pred, average='macro')
    print(f"Accuracy: {acc:.4f}")
    print(f"Macro F1-score: {f1_macro:.4f}\n")
    
    print("Classification Report :")
    print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
    
    plot_confusion_matrix(y_test, y_pred, classes=label_encoder.classes_, title=f'Matrice de confusion - {name.replace("_", " ").title()}')

```
